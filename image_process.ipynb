```python
#
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

im = cv2.imread('2.png')
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
img, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(im, contours, -1, (0,0,255),3 )
plt.imshow(im)
```

```python
#
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

im11 = cv2.imread('11.png')
im22 = cv2.imread('22.png')
im = cv2.imread('black.jpg')

gray = cv2.cvtColor(im11, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
img, contours1, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(im, contours1, -1, (255,255,255), )

gray = cv2.cvtColor(im22, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
img, contours2, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(im, contours2, -1, (255,255,255), 1)

plt.imshow(im)
```

```python
import cv2
import glob as gb
from matplotlib import pyplot as plt
%matplotlib inline

im = cv2.imread('black.jpg')
#Returns a list of all folders with participant numbers
img_path = gb.glob('./picture_process/masks/*.png') 
for path in img_path:
    img  = cv2.imread(path) 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    img, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(im, contours, -1, (255,255,255), )
plt.imshow(im)
```
